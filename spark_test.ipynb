```
import findspark
findspark.init()

from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# 1. Create a Spark Session
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Colab_PySpark_Tutorial") \
    .getOrCreate()

# 2. Create sample data
data = [("Alice", "Sales", 3000),
        ("Bob", "Sales", 4500),
        ("Charlie", "IT", 5000),
        ("Diane", "IT", 6000)]

columns = ["Name", "Department", "Salary"]

# 3. Create DataFrame
df = spark.createDataFrame(data, schema=columns)

# 4. Perform a Transformation: Calculate average salary by department
result = df.groupBy("Department") \
    .agg(F.avg("Salary").alias("Average_Salary"))

# 5. Show results
print("Original Data:")
df.show()

print("Department Averages:")
result.show()

# Stop the session when finished
# spark.stop()
```
